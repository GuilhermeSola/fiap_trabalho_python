##### Correlacao entre preco x atributos numericos por tipo de fabricante.

A seguir vamos fazer uma anlise de como os atributos numéricos se correlacionam com o preco em funcao do fabricante, nos casos em que a amostra nao tem dados sobre o atributo, vamos considerar a correlacao igual a zero (preco nao depende do atributo). Acreditamos que existem fabricantes com caracteristicas paracedas que podem ser agrupados

````python
corr_categ = pd.DataFrame()
for marca in frame.make.unique():
  contagem = frame[(frame.make == marca)].count().min()
  
  if contagem > 1 :
    correlacao = frame[(frame.make == marca)].corr()
    df_corre_categ = DataFrame(correlacao.price).T
    df_corre_categ.insert(loc=0, column='Fabricante', value=marca)
    corr_categ = pd.concat( [corr_categ,df_corre_categ], axis=0 )
    plt.subplots(figsize=(15,1))
    ax=subplot(111)
    sns.heatmap(DataFrame(correlacao.price).T, annot=True, linewidths=.5, ax=ax).set_title("Fabricante : " + marca + " / Amostras : " + str(contagem))
````

Observando as correlacoes temos que a variavel preco varia de forma heterogenia em relacao a distintos fabricantes, ou seja existem grupos de fabricantes com modelos de precos que sao antagonicos. Vamos analisar como dividir os grupos de fabricantes para modelar o preco em funcao dos atributos numéricos.

````python
from sklearn.cluster import KMeans
# Agrupando os tipos de fabricantes semelhantes
dados = DataFrame(corr_categ.replace(np.nan,0))
marca = DataFrame(dados['Fabricante'])
dados = dados.drop(columns=['Fabricante'])
dados = dados.drop(columns=['price'])
kmeans = KMeans(n_clusters=7, random_state=0)
kmeans.fit(dados)
cluster = kmeans.predict(dados)
marca = marca.assign(clusters=cluster)
dados = DataFrame(corr_categ.replace(np.nan,0))
dados = dados.assign(clusters=cluster)

#my_dictionary ={  0: "blue"  , 1: "green",  2 : "red"  }
#my_dictionary ={  0: "blue"  , 1: "green",  2 : "gray" ,  3 : "pink", 4 : "red" }
#my_dictionary ={  0: "blue"  , 1: "green",  2 : "gray" ,  3 : "yellow", 4 : "pink", 5 : "red" }
my_dictionary ={  0: "blue"  , 1: "green",  2 : "gray" ,  3 : "black" , 4 : "yellow", 5 : "pink", 6 : "red" }

cores = list(map(my_dictionary.get, dados["clusters"] ))
dados = dados.assign(cor=cores)

#analisando , 7 clusters identificam a maioria das variaveis 
````

A seguir montamos gráficos para ajudar indentificar o comportamento dos atributos em relacao ao preco do carro por fabricante

````python
def analise_correl_por_categoria(variavel,descricao_variavel,categoria):
  titulo = categoria + " vs |Correl " + variavel + " (" + descricao_variavel + ") " + "/price| "   
  frame = dados[["Fabricante","cor",'clusters',variavel]]
  frame = DataFrame(frame.replace(np.nan,0))
  frame = frame.sort_values(by=['clusters',variavel], ascending=[True,True])
  cores = frame["cor"]
  titulo = categoria + " vs |Correl " + variavel + " (" + descricao_variavel + ") " + "/price| " 
  categorias = frame["Fabricante"]
  fig = sns.barplot(x = variavel, y = categoria, data = frame[["Fabricante",variavel]], palette=cores )
  plt.xlabel("Correl " + variavel + " vs price" )
  plt.ylabel("")
  plt.title(titulo) 
  plt.show(fig)
````

Conforme a leitura anterior, o mais apropriado é que bore e stroke sejam relacionados com o volume da camera de combustao que por sua vez é relacionadaa potencia do carro, vamos montar um gráfico 

````python
# Volume do cilindro = pi^2*altura
# 'compression_ratio' = Volume Total Cilindro/Volume Comprimido 
df['Volume_de_Compressao'] = 3.14*df['stroke']*df['bore']*df['bore']/df['compression_ratio']
df.sort_values(by=['Volume_de_Compressao'])

plt.figure()
sns.regplot(x="Volume_de_Compressao", y="horsepower", data=df[(df["fuel_type"] == "gas") ]) 
sns.regplot(x="Volume_de_Compressao", y="horsepower", data=df[df["fuel_type"] == "diesel"]) 
plt.show()
````

### Análise Numérica de Outliers

Da análise visual de outliers realizada acima vamos utilizar uma métrica de verificação mais apropriada e objetiva afim de criar um patamar aceitável de dados não discrepantes.

Neste projeto vamos considerar [o Método Turco para identificar discrepantes](http://datapigtechnologies.com/blog/index.php/highlighting-outliers-in-your-data-with-the-tukey-method/), que utiliza um __passo de limite__ que é 5 vezes (em nosso projeto) a diferença entre o terceiro (Q3) e o primeiro quartil (Q1). Deste modo, valores que sejam maiores que o Q3 + passo de limite ou menor Q2 - passo de limite sejam sinalizados como outliers.

Construa uma função que receba como parâmetro de entrada um série de dados e exiba os valores discrepantes. Utilize o boilerplate abaixo para completar as instruções faltantes.

*Dica: utilize a função do Numpy de percentil, np.percentile(serie, 25) para quartil 1 e np.percentile(serie, 75) para quartil 3.*

````python
def identificacao_outlier(df, column_name): 
    print("Variavel : ", column_name)
    
    #IMPLEMENTAÇÃO
    limit_step_mult = 5
    q1 = np.percentile(df[column_name], 25)
    q3 = np.percentile(df[column_name], 75)
    
    limit_step = (q3 - q1) * limit_step_mult
    
    outlier_lower = q1 - limit_step
    outlier_upper = q3 + limit_step
    
    print("Limite Inferior : ", outlier_lower, " | Limite Superior : ", outlier_upper)
    print("Lista de Dados Outliers:",list(df[column_name][(df[column_name] < outlier_lower) | (df[column_name] > outlier_upper)]))
    
for i in list(['wheel_base','length','width','height','curb_weight','engine_size','bore','stroke',
             'compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg','price']):
    print("----------------------------------------------------------------------------------------")
    identificacao_outlier(frame_tratado, i)
    
````

### Mapeamento de Dados Paramétricos

Os algoritmos de aprendizado de máquina precisam receber dados que sejam inteiramente numéricos. Dados que representam uma classificação como por exemplo um tipo de carro, como sedan ou hatchback, deve ser convertido em um valor numérico associado, como por exemplo 1 ou 2.

Crie uma função que receba uma lista única e retorne um dicionário com a categoria e um código numérico crescente e incremental para posteriormente utilizarmos como mapeamento.

````python
#IMPLEMENTAÇÃO
def mapear_serie(serie):
    dict_gen = {}
    
    #IMPLEMENTAÇÃO
    i = 0
    s = serie.unique()
    for c in s:
        dict_gen.update({c : i})
        i = i + 1
    print(serie.name, dict_gen)
    return dict_gen
````

Com a funcão criada, crie dicionários para cada coluna paramétrica. Lembre-se que é necessário passar somente valores únicos.

*Dica: utilize a função unique() do dataframe para obter valores únicos de uma determinada série (ou coluna).*

````python
### Criando Variaveis de Dummy
def apply_hot_encoder_columns(df_temp, column):
    for i in range(df_temp[column].min(), df_temp[column].max()):
        df_temp[column+'_'+str(i)] = (df_temp[column] == i).map({True : 1, False : 0})
        
dff2 = frame_tratado
dff2 = dff2.dropna()
dff2["clusters"] = dff2["clusters"].astype(np.int64)
dff2["body_style"] = dff2["body_style"].astype(np.int64)


#################################################################################################

# Retiramos as amostras do cluster zero da base por ter comportamento discrepante em relação ao resto dos dados. ( Vide analise exploratória )

#################################################################################################

dff2 = dff2.drop(dff2[dff2.clusters == 0].index)

dff2['Consumo_Cidade'] = np.log(1/dff2['city_mpg'])


for i in list([ 'clusters','body_style']):
  apply_hot_encoder_columns(dff2,i)

#dff2.head(10).T        
````

Após separarmos os dados adequadamente, selecionar os atributos que irão compor como o modelo deve treinar e qual atributo deve perseguir, o próximo passo é treinar este modelo e verificar, pelos testes, sua performance.

Este estudo também irá levar a escolhermos qual algoritmo devemos utilizar dentre os três selecionados neste ajuste.

Para avaliarmos a performance dos modelos, vamos criar uma função que determinará a pontuação R2.

Não esqueça de avaliar os conjuntos de dados de treino (```X_train```, ```y_train``` e de teste, ```X_test``` e ```y_test```)

````python
from sklearn.metrics import mean_squared_error, r2_score

def pontuacao(modelo, X_test, y_test):
    y_pred = modelo.predict(X_test)
    print("%s R2 score: %.2f" %(modelo,r2_score(y_test, y_pred)))
````

## Validação e Otimização do Modelo

Cada algoritmo de modelo pode oferecer a possibilidade de ajustes de seus parâmetros. Tais ajustes podem contribuir para melhorar ou piorar o modelo, portanto esta fase de otimização e validação é importante entender o patamar de partida, com os valores padrões obtidos nos passos anteriores versus as otimizações.

É comum que as otimizações não sejam capazes de alterar os patamares padrão.

O Scikit Learn oferece uma forma simplificada de testar diversas condições de parâmetros diferentes por meio do ```GridSearchCV```. Após diversos testes é apresentado os parâmetros que obtiveram os melhores resultados.
